## CRAIGSLIST VEHICLE VIN NUMBERS

```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of page to be scraped
url = 'https://austin.craigslist.org/search/cta'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
# print(soup.prettify())
```

```python
# results are returned as an iterable list
results = soup.find_all('li', class_="result-row")
```

```python
links = []
```

```python
# Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return title of listing
        title = result.find('a', class_="result-title").text
        # Identify and return price of listing
        price = result.a.span.text
        # Identify and return link to listing
        link = result.a['href']
        links.append(link)

        # Print results only if title, price, and link are available
        if (title and price and link):
            print('-------------')
            print(title)
            print(price)
            print(link)
    except AttributeError as e:
        print(e)
```

```python
responses = []
```

```python
# get the link and collect them
for url in links:
    print(url)
    # Retrieve page with the requests module
    response = requests.get(url)
    responses.append(response)
```

```python
soups = []
```

```python
# read each link and append them
for res in responses:
    # Create BeautifulSoup object; parse with 'html.parser'
    soup = BeautifulSoup(res.text, 'html.parser')
    soups.append(soup)
```

```python
# len(soups)
```

```python
# Examine the results, then determine element that contains sought info
# print(soup.prettify())
```

```python
results = []
```

```python
# Use CSS class for attributes
for soup in soups:
    # results are returned as an iterable list
    result = soup.find_all('div', class_="mapAndAttrs")
    results.append(result)
```

```python
vins = []
```

```python
# Find the attributes that has the vin numbers

for result in results:
    for x in result:
         #     print(y)
        #     vin = y.find.span.text
        # title = result.find('p', class_="attrgroup").text
        vin = x.find_all('p', class_="attrgroup")[1].text
        # print(type(vin))
        vins.append(vin)
        # print(vin)
        #print(title)
```

```python
vin_nums = []
```

```python
# loop throug the vin list and find the extract the vin numbers
# append it to vin_nums

for i in range(len(mystrings)):
    try:
        for string in mystrings:
            if string[i]=='VIN:':
                vin_nums.append(string[2])
                print(string[2])
    except IndexError:
        print('none')
```

```python
clean_vin = []
```

```python
# Get 17 num standard vin numbers 

for j in vin_nums:
    if len(j) == 17:
        
        clean_vin.append(j)
```
