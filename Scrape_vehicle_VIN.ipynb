## Scrape each url for VIN numbers

```python
import time
import json
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd # import pandas
```

```python
urls_df = pd.read_csv('vehicle_urls/vehicle_urls_df1.csv')
vins = []

# get the link and collect them
for url in urls_df['urls']:
    try:
        time.sleep(2)

        # Retrieve page with the requests module
        response = requests.get(url)
        soup = bs(response.text, 'html.parser')
        result = soup.find('div', class_="mapAndAttrs")
        vin = result.find_all('p', class_="attrgroup")[1].text
        vins.append(vin)
        
    except AttributeError as e:
        print(e)
```

```python
print(vins)
```

```python
len(vins)
```

```python
vin_list = []
for vin in vins:
    vin = vin.replace('\n', ' ').replace('\r', '').split(' ')
    vin_list.append(vin)
```

```python
vin_nums = []
for i in range(len(vin_list)):
    try:
        for vin in vin_list:
            if vin[i]=='VIN:':
                vin_nums.append(vin[2])
                print(vin[2])
    except IndexError:
        print('none')
```

```python
len(vin_nums)
```

```python
# Get the 17-alphanumerical standard vin number
clean_vins = []
for j in vin_nums:
    if len(j) == 17 and j.isalnum():   
        clean_vins.append(j)
```

```python
df = pd.DataFrame(clean_vins, columns=["Vin Number"])
df.head()
```
